# Entregable: Carga y conclusiones del análisis exploratorio de datos.
Haciendo uso de las respuestas que brindaron el comando df.info() podemos identificar cuáles son los campos que comprenden más o menos valores nulos que no serían de utilidad para responder nuestras preguntas objetivo y que podrían representar un sesgo dependiendo del porcentaje en el cuál se encuentren estas variables.

RangeIndex: 891 entries, 0 to 890
Data columns (total 12 columns):
| Nº | Columna  | Valores No Nulos | Tipo de Dato |
|----|----------|------------------|--------------|
| 1  | Survived | 891 non-null     | int64        |
| 5  | Age      | 714 non-null     | float64      |
| 10 | Cabin    | 204 non-null     | str          |
| 11 | Embarked | 889 non-null     | str          |   

Dentro de la tabla se encuentra las cifras de la columna Non-Null Count que verifica cuántas filas y de qué campos no se encuentran con valores nulos. Para el campo de 'Survived' cada una de las entradas de este comprenden valores válidos y aptos para sacar una hipótesis a través de operaciones. Pero los campos de 'Age' 'Cabin' 'Embarked' representan perdidas de datos del 177, 687 y 2 filas comprendiendo una perdida de 19.87%, 77.10% y 0.22% respectivamente (calculado a través de num de nulos entre el num total de filas multiplicado por 100 para obtener el porcentaje). Dentro de un 5% es considerado perdida marginal por lo que podría eliminar esas filas sin introducir un sesgo que descalifique mis conclusiones o imputarlas con medidas de tendencia central ('Embarked'), por otro lado entre 5% y 30% Es necesario la imputación de datos con los metodos anteriormente mencionados o utilizando modelos predictivos ('Age'). Por último, el 77.10% de perdida de información no puede ser solucionado con imputación de datos independientemente de la herramienta que quieras utilizar, donde deberíamos de eliminar la columna porque no es de utilidad para el estudio.

Pasemos a la fase de limpieza.


### Justificación Técnica de las Decisiones de Limpieza

1. **Uso de la Mediana en `Age` en lugar de la Media:**
   Al observar los datos estadísticos previos con `df.describe()`, la distribución de la edad presenta un sesgo debido a la presencia de niños y adultos mayores, desvíandose de una distribución estrictamente normal. La media aritmética es altamente sensible a los valores extremos (outliers), lo que alteraría la estructura real de la población si la usáramos para rellenar los vacíos. La mediana representa el percentil 50 ($28$ años), garantizando una imputación robusta y central no afectada por los extremos.

2. **Nulos de `Embarked` imputados con la moda:**

    Al ser un campo que mantiene un tipo de dato en strings cabe dentro del a categoría de datos nominales donde las respuestas de 'S' 'C' o 'Q' representan puertos y no tienen un sentido matemático. Por lo que la única forma l´+ogica de rellenar los datos faltantes sin alterar la tendencia es usar el puerto donde se subió la gran mayoría de personas. La moda, que en este caso es 'S' por ser el puerto de origen con 644 registros.

3. **Eliminación de la Columna `Cabin`:**
   Con un $77.10\%$ de valores nulos, cualquier intento de imputación tradicional introduciría un ruido masivo y sesgos artificiales en los análisis posteriores, perdiendo utilidad predictiva o explicativa confiable.

```python
# Paso 3: Limpiar
# 1. Imputación de nulos en Age (~20%) con la mediana
mediana_edad = df['Age'].median()
df['Age'] = df['Age'].fillna(mediana_edad)

# 2. Imputación de nulos en Embarked (~2 filas) con la moda
moda_embarque = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(moda_embarque)

# 3. Eliminación de la columna Cabin por exceso de nulos (~77%)
df = df.drop(columns='Cabin')

# 4. Verificación de nulos residuales en las columnas de trabajo
print("Conteo de nulos remanentes por columna:")
print(df.isnull().sum())

# 5. Verificación de filas duplicadas en el dataset
print(f"\nCantidad de registros completamente duplicados: {df.duplicated().sum()}")
```

```python
# Paso 4: Transformar (Feature engineering)
# 1. FamilySize: Creación del tamaño de la familia (Hermanos/Cónyuges + Padres/Hijos + Pasajero)
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# 2. IsAlone: Variable binaria (1 si viaja completamente solo, 0 si viaja acompañado)
df['IsAlone'] = 0
df.loc[df['FamilySize'] == 1, 'IsAlone'] = 1

# 3. AgeGroup: Segmentación de la edad en cuatro rangos ordinales definidos
bins = [0, 12, 18, 60, 100]
labels = ['Niño', 'Adolescente', 'Adulto', 'Mayor']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)

# 4. Title: Extracción del título social del nombre mediante expresiones regulares
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.')

# Mostrar las frecuencias de los títulos extraídos
print("Distribución de Títulos (Title) extraídos de la columna Name:")
print(df['Title'].value_counts())
```

```python
# ==============================================================================
# Paso 5 — Analizar con groupby
# ==============================================================================

print("=== 1. TASA DE SUPERVIVENCIA POR GÉNERO (Sex) ===")
tasa_sex = df.groupby('Sex')['Survived'].mean() * 100
print(tasa_sex.map('{:.2f}%'.format))
print("\n" + "="*50 + "\n")

print("=== 2. TASA DE SUPERVIVENCIA POR CLASE DEL BOLETO (Pclass) ===")
tasa_pclass = df.groupby('Pclass')['Survived'].mean() * 100
print(tasa_pclass.map('{:.2f}%'.format))
print("\n" + "="*50 + "\n")

print("=== 3. TASA DE SUPERVIVENCIA POR GRUPO DE EDAD (AgeGroup) ===")
tasa_age = df.groupby('AgeGroup', observed=False)['Survived'].mean() * 100
print(tasa_age.map('{:.2f}%'.format))
print("\n" + "="*50 + "\n")

print("=== 4. COMBINACIÓN MULTINIVEL: GÉNERO Y CLASE ===")
tabla_sex_pclass = df.groupby(['Sex', 'Pclass'])['Survived'].agg(['mean', 'count'])
tabla_sex_pclass['mean'] = tabla_sex_pclass['mean'] * 100
tabla_sex_pclass.columns = ['Tasa_Supervivencia (%)', 'Total_Pasajeros']
# Usamos display() para mantener el formato de tabla interactiva de Jupyter
display(tabla_sex_pclass.style.format({'Tasa_Supervivencia (%)': '{:.2f}%'}))
print("\n" + "="*50 + "\n")

print("=== 5. ANÁLISIS MULTIVARIADO AVANZADO (Hipótesis) ===")
analisis_multivariado = df.groupby(['Sex', 'Pclass', 'AgeGroup'], observed=False)['Survived'].agg(['mean', 'count'])
analisis_multivariado.columns = ['Tasa_Supervivencia (%)', 'Total_Pasajeros']
analisis_multivariado['Tasa_Supervivencia (%)'] = analisis_multivariado['Tasa_Supervivencia (%)'] * 100

# Ordenar de forma descendente y aplicar formato de tabla estilizada
analisis_final = analisis_multivariado.sort_values(by='Tasa_Supervivencia (%)', ascending=False)
display(analisis_final.style.format({'Tasa_Supervivencia (%)': '{:.2f}%'}, na_rep='0.00%'))
```
