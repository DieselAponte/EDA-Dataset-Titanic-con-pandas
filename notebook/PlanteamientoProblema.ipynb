La variable objetivo es la **supervivencia** de caracter binario, ya que representa el resultado que se desea explicar o predecir. Quisiera realizar un análisis multivariado contemplando una relación principalmente la clase del boleto (`Pclass`), la edad (`Age`) y el género (`Sex`) para verificar su impacto en la probabilidad de supervivencia.
